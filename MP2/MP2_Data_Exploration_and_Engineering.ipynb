#### 6 Plot diagrams that visualize the differences in red and white wine samples

```python
columns_compare = ['quality', 'alcohol', 'fixed acidity']
```

```python
# Loop through each column and plot the histograms for red and white wines
for col in columns_compare:
    plt.figure(figsize=(8, 6))
    
    plt.hist(dred[col], bins=10, alpha=0.5, label='Red Wine', color='red')
    
    plt.hist(dwhite[col], bins=10, alpha=0.5, label='White Wine', color='blue')
    
    plt.title(f'Comparison of {col} between red and white wine')
    plt.xlabel(col)
    plt.ylabel('Amount')
    plt.legend()
    
    plt.show()
```

```python
from scipy import stats
z = np.abs(stats.zscore(datawine['residual sugar']))
print(z)
```

```python
lower_threshold_z = -3
upper_threshold_z = 3

outlier_indices = np.where((z > upper_threshold_z) | (z < lower_threshold_z ))[0]
outlier_index_labels = datawine.iloc[outlier_indices].index
no_outliers = datawine.drop(outlier_index_labels)
print("Original DataFrame:", datawine.shape)
print("DataFrame after Removing Outliers:", no_outliers.shape)
```
